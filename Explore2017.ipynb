```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_raw = pd.read_csv('survey_results_public.csv', index_col='Respondent')
df = df_raw[df_raw['Salary'].notnull()]
```

```python
for col in df.columns:
    print(col)
```

```python
pd.DataFrame.hist(df, column='Salary')
```

```python
df_us = df[df['Country'] == 'United States']
ax = plt.gca()
kwargs = {'column':'Salary', 'ax':ax}
pd.DataFrame.hist(df_us[df_us['Gender'] == 'Male'], label='Men', **kwargs)
pd.DataFrame.hist(df_us[df_us['Gender'] == 'Female'], label='Women', **kwargs)
plt.legend()
```

```python
ax = plt.gca()
kwargs = {'column':'Salary', 'ax':ax}
pd.DataFrame.hist(df_us[df_us['ChangeWorld'] == 'Somewhat agree'], label='Somewhat agree', **kwargs)
pd.DataFrame.hist(df_us[df_us['ChangeWorld'] == 'Strongly agree'], label='Strong agree', **kwargs)
pd.DataFrame.hist(df_us[df_us['ChangeWorld'] == 'Disagree'], label='Diagree', **kwargs)
plt.legend()
```

```python
ax = plt.gca()
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '3 to 4 years'], label='3-4', **kwargs)
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '2 to 3 years'], label='2-3', **kwargs)
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '1 to 2 years'], label='1-2', **kwargs)
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == 'Less than a year'], label='<1', **kwargs)
plt.title('Salary by Years of Job Experience')
plt.legend()
```

```python
ax = plt.gca()
kwargs = {'column':'Salary', 'ax':ax, 'alpha':0.7}
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '20 or more years'], label='20+', **kwargs)
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '2 to 3 years'], label='2-3', **kwargs)
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '7 to 8 years'], label='7-8', **kwargs)
pd.DataFrame.hist(df_us[df_us['YearsCodedJob'] == '14 to 15 years'], label='14-15', **kwargs)
plt.title('Salary by Years of Job Experience')
plt.legend()
```

```python
df['MajorUndergrad'].unique()
```

```python
ax = plt.gca()
kwargs = {'column':'Salary', 'ax':ax}
major_col = 'MajorUndergrad'
cs_filter = ((df_us[major_col] == 'Computer science or software engineering') |
             (df_us[major_col] == 'Computer engineering or electrical/electronics engineering') |
             (df_us[major_col] == 'Computer programming or Web development'))
eng_filter = df_us[major_col] == 'A non-computer-focused engineering discipline'
other_filter = (~cs_filter) & (~eng_filter) & df_us['MajorUndergrad']
pd.DataFrame.hist(df_us[cs_filter], label='Computers', **kwargs)
pd.DataFrame.hist(df_us[other_filter], label='Other', **kwargs)
pd.DataFrame.hist(df_us[eng_filter], label='Engineering', **kwargs)
plt.title('Salary by Undergraduate Major')
plt.legend()
```
